```python
with open("input.txt") as f:
    input = f.read().splitlines()
    input = {(i, j): c for i, l in enumerate(input) for j, c in enumerate(l)}
```

```python
def dfs(x, y, target, group):
    if (x, y) in group:
        return
    if input.get((x, y)) != target:
        return
    group.add((x, y))
    dfs(x+1, y, target, group)
    dfs(x-1, y, target, group)
    dfs(x, y+1, target, group)
    dfs(x, y-1, target, group)
    
```

```python
explored = set()
groups = []
for elem in input:
    if elem in explored:
        continue
    group = set()
    dfs(*elem, input[elem], group)
    groups.append(group)
    explored.update(group)
```

```python
total_cost = 0
for g in groups:
    perimeter = 0
    for elem in g:
        for i in [(-1, 0), (0, -1), (1, 0), (0, 1)]:
            if (elem[0] + i[0], elem[1] + i[1]) not in g:
                perimeter += 1
    print(input[elem], len(g), perimeter)
    total_cost += len(g) * perimeter
print(f"total_cost = {total_cost}")
```

## Part 2

```python
def corner_detection(x, y, group):
    n_corners = 0
    for dx, dy in [(-1, -1), (-1, 1), (1, -1), (1, 1)]:
        e1 = (x+dx, y) in group
        e2 = (x, y+dy) in group
        if (x+dx, y+dy) not in group:
            if e1 == e2:
                n_corners += 1
        elif not e1 and not e2:
            n_corners += 1
    return n_corners
```

```python
total_cost = 0
for g in groups:
    for elem in g:
        total_cost += corner_detection(*elem, g) * len(g)
print(f"total_cost = {total_cost}")
```
