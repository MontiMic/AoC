```python
with open("input.txt") as f:
    input = f.readlines()
    input = [x.strip() for x in input]
```

```python
count = 0
significative = []
for x, line in enumerate(input):
    for y, c in enumerate(line):
        if y + 4 <= len(line) and line[y:y+4] in ['XMAS', 'SAMX']:
            count += 1
            for i in range(4):
                significative.append((x, y+i))
        if x + 3 < len(input) and input[x][y] + input[x+1][y] + input[x+2][y] + input[x+3][y] in ['XMAS', 'SAMX']:
            count += 1 
            for i in range(4):
                significative.append((x+i, y))
        if x + 3 < len(input) and y + 3 < len(line) and input[x][y] + input[x+1][y+1] + input[x+2][y+2] + input[x+3][y+3] in ['XMAS', 'SAMX']:
            count += 1
            for i in range(4):
                significative.append((x+i, y+i))
        if x + 3 < len(input) and y - 3 >= 0 and input[x][y] + input[x+1][y-1] + input[x+2][y-2] + input[x+3][y-3] in ['XMAS', 'SAMX']:
            count += 1
            for i in range(4):
                significative.append((x+i, y-i))
count
```

```python
filtered_input = []
for x, line in enumerate(input):
    new_line = list(line)
    for y, c in enumerate(line):
        if (x, y) not in significative:
            new_line[y] = '.'
    filtered_input.append(''.join(new_line))
for line in filtered_input:
    print(line)
```
